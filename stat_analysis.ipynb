```python
import pandas as pd

# Load the cleaned dataset
df = pd.read_csv('../machine_learning_2025/datasets/pima_indian_diabetes_dataset/cleaned_dataset.csv')

# Print min, max, mean, and standard deviation for all columns
print(df.describe().loc[['min', 'max', 'mean', 'std']])
```

```python
import matplotlib.pyplot as plt

# Graficar cada columna numrica en grficas separadas
for col in df.columns:
    plt.figure()
    df[col].plot(kind='hist', bins=30, alpha=0.7)
    plt.title(f'Distribucin de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.grid(True)
    plt.show()
```
